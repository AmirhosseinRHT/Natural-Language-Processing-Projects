برای این کار چند راهبرد مختلف بر پایه‌ی مدل‌های زبانی حرف‌محور از نوع <b>n-gram</b> پیاده‌سازی می‌کنید. هدف نهایی این است که مدلی بسازید که با کمترین تعداد حدس اشتباه، واژه‌های مخفی را بازسازی کند.

برای ساده‌تر شدن اجرا، فایل واژه‌های موردنیاز همراه تکلیف قرار داده شده است و در این نسخه به هیچ دانلود بیرونی یا کتابخانه‌ی اضافی برای داده نیاز ندارید.

# <div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;">سامانه‌ی رمزگشایی سیگنال (۷ نمره)</div>

<div dir="rtl" style="text-align: right; line-height: 2; font-family: Vazir, Tahoma, sans-serif;">
<b>توضیح سناریو:</b> فرض کنید یک کانال ارتباطی قدیمی فقط یک واژه‌ی کلیدی را مخابره می‌کند، اما بخشی از نویز کانال باعث شده فقط جایگاه حروف مشخص باشد و خود حروف پنهان شوند. شما در هر مرحله تنها مجاز هستید <b>یک حرف</b> پیشنهاد دهید. اگر حرف درست باشد، تمام رخدادهای آن در واژه آشکار می‌شود. اگر غلط باشد، یکی از فرصت‌های شما کم می‌شود. بازی وقتی تمام می‌شود که یا کل واژه بازیابی شود، یا تعداد خطاها از حد مجاز عبور کند.

در ادامه، یک پیاده‌سازی ساده از این فرایند آمده است. <b>در این بخش نیازی به پیاده‌سازی ندارید.</b>
</div>

```python
def decode_signal(secret_word, decoder, max_errors=8, verbose=True, **decoder_args):
    """
    Run the masked-word decoding game and return the number of wrong guesses.

    secret_word: target word in lowercase letters
    decoder: function that proposes the next character
        inputs:
            mask: current known pattern, with "_" for unknown positions
            guessed: set of characters already proposed
            decoder_args: optional extra keyword arguments
    max_errors: maximum number of wrong guesses allowed
    verbose: whether to print step-by-step details
    decoder_args: extra keyword arguments passed to the decoder
    """
    secret_word = secret_word.lower()
    mask = ["_"] * len(secret_word)
    guessed = set()

    if verbose:
        print("Signal decoding started")
        print("Target pattern:", " ".join(mask), "| Word length:", len(secret_word))

    errors = 0

    while errors < max_errors:
        if verbose:
            print("Remaining allowed errors:", max_errors - errors)

        guess = decoder(mask, guessed, **decoder_args)

        if verbose:
            print("Proposed character:", guess)

        if guess in guessed:
            if verbose:
                print("This character was already tried.")
            errors += 1
        else:
            guessed.add(guess)

            if guess in secret_word and len(guess) == 1:
                for index, char in enumerate(secret_word):
                    if char == guess:
                        mask[index] = char
                if verbose:
                    print("Correct guess:", " ".join(mask))
            else:
                if len(guess) != 1:
                    print("Please enter exactly one character.")
                if verbose:
                    print("Wrong guess.")
                errors += 1

        if "_" not in mask:
            if verbose:
                print("Decoding completed successfully.")
            return errors

    if verbose:
        print("No attempts left. The original word was:", secret_word)
    return errors
```

### <div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;">سؤال ۱ (۱ نمره)</div>

<div dir="rtl" style="text-align: right; line-height: 2; font-family: Vazir, Tahoma, sans-serif;">
در این نسخه، یک فایل داده‌ی متنی با نام <code>hw1_word_bank.txt</code> همراه تکلیف قرار داده شده است. در هر خط از این فایل یک واژه‌ی انگلیسی آمده است. در این بخش باید واژه‌ها را از فایل بخوانید و <b>نوع‌های واژگانی یکتا</b> را به‌دست آورید، با این شرط که فقط واژه‌هایی را نگه دارید که کاملاً از حروف الفبا تشکیل شده باشند. همه‌ی واژه‌ها باید به حروف کوچک تبدیل شوند.

سپس این مجموعه‌ی واژه‌ها را به‌صورت تصادفی با <code>random.shuffle</code> به‌هم بزنید و آن را به دو بخش جدا تقسیم کنید:
<ul>
  <li><b>مجموعه‌ی ارزیابی</b>: دقیقاً 1000 واژه</li>
  <li><b>مجموعه‌ی آموزش</b>: بقیه‌ی واژه‌ها</li>
</ul>

دقت کنید که واژه‌های مجموعه‌ی ارزیابی نباید در مجموعه‌ی آموزش باشند. هدف این است که عامل شما روی واژه‌هایی سنجیده شود که مستقیماً آن‌ها را در آموزش ندیده است.

<b>وظیفه:</b> دو لیست <code>train_words</code> و <code>eval_words</code> را بسازید.<br>
<b>بررسی:</b> از assertionهای پایین استفاده کنید.
</div>

```python
from pathlib import Path
import random
import re

random.seed(1)

WORD_BANK_PATH = Path("hw1_word_bank.txt")
if not WORD_BANK_PATH.exists():
    raise FileNotFoundError(
        "Could not find 'hw1_word_bank.txt'. Put the data file next to this notebook."
    )

# train_words stores word types used for training
train_words = []

# eval_words stores 1000 word types used for evaluation
eval_words = []


with open(WORD_BANK_PATH, 'r', encoding='utf-8') as f:
    words = {line.strip().lower() for line in f if line.strip().isalpha()}

unique_words = list(words)
random.shuffle(unique_words)

eval_words = unique_words[:1000]
train_words = unique_words[1000:]


print("Number of word types in eval =", len(eval_words))
print("Number of word types in train =", len(train_words))
```

### <div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;">سؤال ۲ (۰.۵ نمره)</div>

<div dir="rtl" style="text-align: right; line-height: 2; font-family: Vazir, Tahoma, sans-serif;">
برای داشتن یک خط مبنا، ابتدا یک روش بسیار ساده و کاملاً تصادفی پیاده‌سازی می‌کنید. این روش باید از میان حروف <code>a</code> تا <code>z</code>، یکی را به‌صورت تصادفی انتخاب کند؛ البته حروفی که قبلاً حدس زده شده‌اند نباید دوباره انتخاب شوند.

برای ارزیابی کیفیت این روش و روش‌های بعدی، تابع <code>evaluate_decoder</code> در اختیار شما قرار داده شده است. این تابع میانگین تعداد حدس‌های اشتباه را روی تمام واژه‌های مجموعه‌ی ارزیابی محاسبه می‌کند.

<b>وظیفه:</b> تابع <code>random_decoder</code> را کامل کنید.<br>
<b>بررسی:</b> از assertionهای پایین استفاده کنید.
</div>

```python
def evaluate_decoder(decoder, evaluation_words):
    """
    Measure the average number of wrong guesses made by a decoder
    over all words in the evaluation set.
    """
    total_errors = 0
    for word in evaluation_words:
        total_errors += decode_signal(word, decoder, 26, False)
    return total_errors / float(len(evaluation_words))
```

```python
import string

def random_decoder(mask, guessed, **kwargs):
    letters = string.ascii_lowercase
    candidates = [ch for ch in letters if ch not in guessed]
    return random.choice(candidates) if candidates else None

sample_word = random.choice(eval_words)
print("Selected word =", sample_word)
print("Mistakes made by random decoder =", decode_signal(sample_word, random_decoder, 26, False))

result = evaluate_decoder(random_decoder, eval_words)
print("\nEvaluating the random decoder on the full evaluation set")
print("Average number of incorrect guesses:", result)
```

<div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;"><b>برای آزمون شما:</b></div>

```python
assert(result > 10 and result < 20)
```

### <div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;">سؤال ۳ (۱ نمره)</div>

<div dir="rtl" style="text-align: right; line-height: 2; font-family: Vazir, Tahoma, sans-serif;">
اکنون اولین مدل واقعی خود را می‌سازید: یک <b>مدل زبانی یونی‌گرام</b> در سطح حرف. در این مدل، باید فراوانی هر حرف را در کل واژه‌های آموزشی محاسبه کنید. سپس حدس‌زن شما باید حرفی را برگرداند که بر اساس این توزیع، بیشترین احتمال را دارد؛ البته حروفی که قبلاً حدس زده شده‌اند باید کنار گذاشته شوند.

<b>وظیفه:</b> فراوانی حروف را در متغیر <code>char_unigram_counts</code> ذخیره کنید و سپس تابع <code>unigram_decoder</code> را کامل کنید.<br>
<b>بررسی:</b> از assertionهای پایین استفاده کنید.
</div>

```python
char_unigram_counts = None

char_unigram_counts = {letter: 0 for letter in string.ascii_lowercase}

for word in train_words:
    for char in word:
        if char in char_unigram_counts:
            char_unigram_counts[char] += 1


def unigram_decoder(mask, guessed, char_unigram_counts=char_unigram_counts):

    best_letter = None
    best_count = -1
    
    for letter in string.ascii_lowercase:
        if letter not in guessed:
            if char_unigram_counts[letter] > best_count:
                best_count = char_unigram_counts[letter]
                best_letter = letter
    
    return best_letter if best_letter else random.choice([ch for ch in string.ascii_lowercase if ch not in guessed])


result = evaluate_decoder(unigram_decoder, eval_words)
print("Evaluating the unigram decoder on the full evaluation set")
print("Average number of incorrect guesses:", result)
```

<div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;"><b>برای آزمون شما:</b></div>

```python
assert(result > 5 and result < 15)
```

### <div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;">سؤال ۴ (۱ نمره)</div>

<div dir="rtl" style="text-align: right; line-height: 2; font-family: Vazir, Tahoma, sans-serif;">
طول واژه می‌تواند یک سرنخ مفید باشد. برای مثال، توزیع حروف در واژه‌های کوتاه و بلند دقیقاً یکسان نیست. در این بخش باید یک مدل یونی‌گرام بسازید که به <b>طول واژه</b> شرطی شده باشد؛ یعنی برای هر طول، یک توزیع جداگانه از حروف داشته باشید.

در زمان آزمون ممکن است با طولی روبه‌رو شوید که در آموزش ندیده‌اید. در این حالت، مدل شما باید به مدل یونی‌گرام ساده‌ی سؤال قبل برگردد.

<b>وظیفه:</b> فراوانی حروف شرطی‌شده بر طول را در <code>char_unigram_by_length</code> ذخیره کنید و سپس تابع <code>length_aware_unigram_decoder</code> را کامل کنید.<br>
<b>بررسی:</b> از assertionهای پایین استفاده کنید.
</div>

```python
char_unigram_by_length = {}

for word in train_words:
    length = len(word)
    if length not in char_unigram_by_length:
        char_unigram_by_length[length] = {letter: 0 for letter in string.ascii_lowercase}
    
    for char in word:
        if char in char_unigram_by_length[length]:
            char_unigram_by_length[length][char] += 1


def length_aware_unigram_decoder(
    mask,
    guessed,
    char_unigram_by_length=char_unigram_by_length,
    char_unigram_counts=char_unigram_counts,
):
    word_length = len(mask)
    
    if word_length in char_unigram_by_length:
        freq_dict = char_unigram_by_length[word_length]
    else:
        freq_dict = char_unigram_counts
    
    best_letter = None
    best_count = -1
    
    for letter in string.ascii_lowercase:
        if letter not in guessed:
            if freq_dict[letter] > best_count:
                best_count = freq_dict[letter]
                best_letter = letter
    
    if best_letter is None:
        candidates = [ch for ch in string.ascii_lowercase if ch not in guessed]
        return random.choice(candidates) if candidates else None
    
    return best_letter

result = evaluate_decoder(length_aware_unigram_decoder, eval_words)
print("Evaluating the length-aware unigram decoder on the full evaluation set")
print("Average number of incorrect guesses:", result)
```

<div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;"><b>برای آزمون شما:</b></div>

```python
assert(result > 5 and result < 15)
```

### <div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;">سؤال ۵ (۱ نمره)</div>

<div dir="rtl" style="text-align: right; line-height: 2; font-family: Vazir, Tahoma, sans-serif;">
در این بخش باید یک <b>مدل بی‌گرام</b> روی حروف بسازید. ترتیب حروف در واژه بسیار مهم است و مدل‌های قبلی این موضوع را در نظر نمی‌گرفتند. برای نمونه، اگر بخشی از الگو به صورت <code>n _ s s</code> باشد، احتمال اینکه جای خالی با حرف <code>e</code> پر شود زیاد است.

مدل بی‌گرام شما باید برای هر جای خالی از <b>حرف سمت چپ</b> به عنوان متن زمینه استفاده کند. برای مثال، در الگوی <code>e _ c _ b _ _</code> برای سه جای خالی اول، متن زمینه‌ی چپ مشخص است؛ اما برای آخرین جای خالی چنین چیزی نداریم. در این حالت می‌توانید به مدل یونی‌گرام برگردید.

هنگام آموزش، ابتدای واژه را هم درست مدیریت کنید؛ مثلاً با یک نماد ویژه مانند <code>$</code>. اینکه آیا به نماد پایان واژه هم نیاز دارید یا نه، بخشی از طراحی شماست.

<b>وظیفه:</b> شمارش‌های بی‌گرام را در <code>char_bigram_counts</code> ذخیره کنید و سپس تابع <code>bigram_decoder</code> را کامل کنید.
</div>

```python
char_bigram_counts = {}

for word in train_words:
    augmented_word = '$' + word
    
    for i in range(len(augmented_word) - 1):
        bigram = (augmented_word[i], augmented_word[i+1])
        if bigram not in char_bigram_counts:
            char_bigram_counts[bigram] = 0
        char_bigram_counts[bigram] += 1


def bigram_decoder(mask, guessed, char_bigram_counts=char_bigram_counts):
    
    empty_positions = [i for i, ch in enumerate(mask) if ch == '_']
    
    if not empty_positions:
        return None

    candidate_scores = {}
    
    for letter in string.ascii_lowercase:
        if letter in guessed:
            continue
        
        total_score = 0
        positions_considered = 0
        
        for pos in empty_positions:
            if pos > 0:
                left_char = mask[pos - 1]
                if left_char != '_':
                    bigram = (left_char, letter)
                    score = char_bigram_counts.get(bigram, 0)
                    total_score += score
                    positions_considered += 1
                else:
                    pass
            else:
                bigram = ('$', letter)
                score = char_bigram_counts.get(bigram, 0)
                total_score += score
                positions_considered += 1
        
        if positions_considered == 0:
            total_score = char_unigram_counts.get(letter, 0)
        else:
            total_score = total_score / positions_considered
        
        candidate_scores[letter] = total_score
    
    if not candidate_scores:
        return None
    
    best_letter = max(candidate_scores, key=candidate_scores.get)
    return best_letter


result = evaluate_decoder(bigram_decoder, eval_words)
print("Evaluating the bigram decoder on the full evaluation set")
print("Average number of incorrect guesses:", result)
```

### <div dir="rtl" style="text-align: right; font-family: Vazir, Tahoma, sans-serif;">سؤال ۶ (۲ نمره)</div>

<div dir="rtl" style="text-align: right; line-height: 2; font-family: Vazir, Tahoma, sans-serif;">
اکنون نوبت طراحی یک عامل بهتر است. در این بخش می‌توانید خلاقانه‌تر عمل کنید. برای مثال:
<ul>
  <li>شرطی‌سازی بهتر بر طول واژه</li>
  <li>روش‌های هموارسازی (smoothing)</li>
  <li>مدل‌های n-gram با مرتبه‌ی بالاتر</li>
  <li>مدل‌های دوطرفه</li>
</ul>

نمره‌ی این بخش بر اساس عملکرد مدل شما روی یک مجموعه‌ی آموزش/آزمون از پیش‌ساخته‌شده محاسبه می‌شود. اگر <code>x</code> میانگین تعداد خطاها باشد:
<ul>
  <li><b>۲ نمره</b> اگر <code>x &lt; 6.0</code></li>
  <li><b>۱.۵ نمره</b> اگر <code>6.0 &lt;= x &lt; 6.5</code></li>
  <li><b>۱ نمره</b> اگر <code>6.5 &lt;= x &lt; 7.0</code></li>
  <li><b>۰.۵ نمره</b> اگر <code>7.0 &lt;= x &lt; 8.0</code></li>
  <li><b>۰ نمره</b> اگر <code>x &gt;= 8.0</code></li>
</ul>

<b>نکته‌ها:</b>
<ul>
  <li>برای درک بهتر رفتار مدل، می‌توانید چند تقسیم‌بندی متفاوت از داده را امتحان کنید.</li>
  <li>کد شما باید در زمان معقول اجرا شود؛ پیاده‌سازی‌های خیلی کند برای این تکلیف مناسب نیستند.</li>
</ul>

<b>وظیفه:</b> تابع <code>my_best_decoder</code> را کامل کنید.
</div>

```python
def get_length_aware_probs(word_length):
    """Return letter probabilities conditioned on word length with add-1 smoothing."""
    probs = {}
    if word_length in char_unigram_by_length:
        total = sum(char_unigram_by_length[word_length].values())
        for letter in string.ascii_lowercase:
            count = char_unigram_by_length[word_length].get(letter, 0)
            probs[letter] = (count + 1) / (total + 26)
    else:
        total = sum(char_unigram_counts.values())
        for letter in string.ascii_lowercase:
            count = char_unigram_counts.get(letter, 0)
            probs[letter] = (count + 1) / (total + 26)
    return probs


def get_bigram_prob(prev_char, curr_char, length_probs):
    """Return bigram probability with Good-Turing discount and fallback."""
    bigram = (prev_char, curr_char)
    bigram_count = char_bigram_counts.get(bigram, 0)
    total_from_prev = sum(1 for bg in char_bigram_counts if bg[0] == bigram[0])
    
    if total_from_prev == 0:
        return length_probs.get(curr_char, 0.01)
    
    prob = max(0, bigram_count - 0.5) / total_from_prev
    return prob + 0.01


def my_best_decoder(mask, guessed):
    
    word_length = len(mask)
    length_probs = get_length_aware_probs(word_length)
    empty_positions = [i for i, ch in enumerate(mask) if ch == '_']
    
    if not empty_positions:
        return None
    
    scores = {}
    
    for letter in string.ascii_lowercase:
        if letter in guessed:
            continue
        
        total_score = 0
        positions_processed = 0
        
        for pos in empty_positions:
            position_score = 1.0
            
            # Left bigram context
            if pos > 0 and mask[pos - 1] != '_':
                left_char = mask[pos - 1]
                position_score *= get_bigram_prob(left_char, letter, length_probs)
                positions_processed += 1
            
            # Right bigram context  
            if pos < word_length - 1 and mask[pos + 1] != '_':
                right_char = mask[pos + 1]
                bigram_rev = (letter, right_char)
                rev_count = char_bigram_counts.get(bigram_rev, 0)
                total_from_letter = sum(1 for bg in char_bigram_counts if bg[0] == letter)
                
                if total_from_letter > 0:
                    rev_prob = rev_count / total_from_letter
                else:
                    rev_prob = length_probs.get(right_char, 0.01)
                
                position_score *= (rev_prob + 0.01)
                positions_processed += 1
            
            # Fallback
            if positions_processed == 0:
                position_score = length_probs[letter]
                positions_processed = 1
            
            # Position bonuses
            if pos == 0:
                position_score *= 1.3
            elif pos == word_length - 1:
                position_score *= 1.2
            
            total_score += position_score
        
        scores[letter] = total_score / positions_processed
    
    if not scores:
        candidates = [ch for ch in string.ascii_lowercase if ch not in guessed]
        return random.choice(candidates) if candidates else None
    
    return max(scores, key=scores.get)

# Evaluate
result = evaluate_decoder(my_best_decoder, eval_words)
print("Average number of incorrect guesses:", result)
```
